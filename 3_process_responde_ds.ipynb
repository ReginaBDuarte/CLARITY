```python
import pandas as pd
import numpy as np
import json
import ast

```

```python
#fatores de risco:


translate_map = {
    "Age": "Idade",
    "Gender": "Género",
    "Smoking": "Tabagismo",
    "Arterial Hypertension": "Hipertensão Arterial",
    "Systolic Blood Pressure": "Pressão Arterial Sistólica",
    "Diastolic Blood Pressure": "Pressão Arterial Diastólica",
    "Dyslipidemia": "Dislipidemia",
    "Total Cholesterol": "Colesterol Total",
    "HDL-Cholesterol": "Colesterol HDL",
    "Triglycerides": "Triglicéridos",
    "LDL-Cholesterol": "Colesterol LDL",
    "Diabetes Mellitus": "Diabetes Mellitus",
    "non-HDL Cholesterol": "Colesterol não HDL"
}

translate_modificadores = {
    "Coronary Calcium Score > 0": "Score de Cálcio elevado",
    "Calcium Score = 0": "Score de Cálcio = 0",
    "Carotid Plaques": "Placas Carótidas",
    "Obesity": "Obesidade",
    "Pre-Diabetes": "Pre-Diabetes",
    "Family History of atherosclerotic cardiovascular disease (ASCVD)": "História Familiar de doença cardiovascular aterosclerótica",
    "Elevated Lp(a)": "Lp (a) aumentada",
    "Increased Arterial Stiffness": "Rigidez Arterial aumentada",
    "Elevated High-sensitivity CRP": "PCR alta sensibilidade elevada",
    "Chronic Inflammatory Disease": "Doença Inflamatória Crónica",
    "HIV Infection": "Infecção HIV",
    "Obstructive Sleep Apnea Syndrome": "Sindrome Apneia Obstrutiva do Sono",
    "Non-Alcoholic Fatty Liver Disease": "Esteatose Hepática Não Alcoólica",
    "Erectile Dysfunction": "Disfunção erétil",
    "Premature Menopause": "Menopausa precoce",
    "Past history of Pregnancy Complications (Preeclampsia, Hypertension, or Diabetes)":
        "Antecedentes de Gravidez com complicações (Pre-eclampsia, Hipertensão arterial ou diabetes)",
    "Chronic Obstructive Pulmonary Disease (COPD)": "Doença pulmonar obstrutiva crónica (DPOC)",
    "Cancer": "Cancro",
    "Other": "Outro",
    'Outro (especificar)': 'Outro',
    'Other (specify)': 'Outro'

}



def translate_keys(d):
    return {translate_map.get(k, k): v for k, v in d.items()}

def translate_mod_keys(d):
    return {translate_modificadores.get(k, k): v for k, v in d.items()}



```

```python
data = pd.read_csv('model_responses_5.csv', sep=';')
data.head()
```

```python
data['fatores_de_risco'] = data['fatores_de_risco'].apply(ast.literal_eval)
data['fatores_de_risco'] = data['fatores_de_risco'].apply(translate_keys)
data_fac = data['fatores_de_risco'].apply(pd.Series)

data['modificadores_de_risco'] = data['modificadores_de_risco'].apply(ast.literal_eval)
data['modificadores_de_risco'] = data['modificadores_de_risco'].apply(translate_mod_keys)
data_mod = data['modificadores_de_risco'].apply(pd.Series)

# Junta ao datafram
data = pd.concat([data.drop(columns=['fatores_de_risco']), data_fac], axis=1)
data = pd.concat([data.drop(columns=['modificadores_de_risco']), data_mod], axis=1)


data.head()
```
